# Gaussian Processes

```python
import numpy as np
import matplotlib.pyplot as plt
import pdb
```

### Gaussian Process Prior
$\large f \sim \mathcal{GP}(\mathbf{0},\mathbf{K})$

```python
# test locations
N = 50
X = np.linspace(-3,3,N).reshape(N,1)
```

#### Kernel function

$\mathbb{C}ov(f(x_1),f(x_2) = \mathbf{K}(x_1,x_2)$

```python
def gauss_kern(x1,x2):
    try:
        return np.exp(-0.5*np.matmul((x1-x2).T,(x1-x2)))
    except ValueError:
        print("1:",x1," 2:",x2)
        pdb.set_trace()

def build_K(X,Y):
    N = X.shape[0]
    M = Y.shape[0]
    K = np.zeros((N,M))
    for i in range(N):
        for j in range(M):
            K[i,j] = gauss_kern(X[i],Y[j])
    return K
```

```python
K = build_K(X,X)
```

```python
_ = plt.imshow(K)
```

### Sampling from the prior

```python
# sampling from the prior
prior_mean = np.zeros_like(X).flatten() # (N,) - mean must be one dimensional?
num_samples = 30

f  = np.random.multivariate_normal(mean=prior_mean, cov=K,size=num_samples)
```

```python
fig = plt.figure(figsize=(14,6))
for sample in f:
    _ = plt.plot(X,sample,'C0')
plt.xlabel("X")
plt.ylabel("F(X)")
```

## Noise-_free_ Regression

```python
# generate data
X = np.array([[0,np.pi/4.0,5*np.pi/7.0,np.pi,3*np.pi/2.0]]).reshape((-1,1))
F = np.sin(X)

# ground truth + test points
x_ = np.linspace(0,2*np.pi).reshape(-1,1)
f_ = np.sin(x_)
```

```python
_ = plt.plot(x_,f_)
_ = plt.plot(X,F,'C1x')
```

### Conditioning
$p(f(x^*)|f(X),x^*,X) = \mathcal{GP}(\mu,\nu)\quad$ where:

- $\mu(x^*) = k(x^*,\mathbf{X})K^{-1}\mathbf{f}$
- $\nu(x^*,x^*) = k(x^*,x^*)-k(x^*,\mathbf{X})K^{-1}k(\mathbf{X},x^*)$


```python
# mean
k_x_star_X = build_K(x_,X)
k_X_x_star = build_K(X,x_)
k_x_star_x_star = build_K(x_,x_)
Kxx = build_K(X,X)

mu = k_x_star_X @ np.linalg.solve(Kxx,F)
```

```python
# covariance matrix
cov = k_x_star_x_star - k_x_star_X @ np.linalg.solve(Kxx,k_X_x_star)
```

```python
# sample from GP posterior
num_samples = 50
samples = np.random.multivariate_normal(mean=mu.flatten(),cov=cov,size=num_samples)
```

```python
fig = plt.figure(figsize=(14,6))
_ =  plt.plot(x_,f_,'C1')
for sample in samples:
    _ = plt.plot(x_,sample,'C0',alpha=0.4)
_ =  plt.plot(x_,f_,'C1')
plt.xlabel("X")
plt.ylabel("F(X)")
plt.legend(["sin(x)","samples"])
```

## Regression with output noise

Now assume we can't directly observe function values $f_i$'s for inputs $x_i$'s, instead we observe noise corrupted versions of the $f_i$'s. In other words, we observe $y = f(x)+\varepsilon$, where $\varepsilon \sim \mathcal{N}(0,\sigma^2)$

```python
# generate data
noise_var = 0.05
X = np.array([[0,np.pi/4.0,5*np.pi/7.0,np.pi,3*np.pi/2.0]]).reshape((-1,1))
Y = np.sin(X) + np.random.normal(scale=noise_var,size=X.shape)

# ground truth + test points
x_ = np.linspace(0,2*np.pi).reshape(-1,1)
f_ = np.sin(x_)
```

```python
_ = plt.plot(x_,f_)
_ = plt.plot(X,Y,'C1x')
```

### Conditioning
$p(f(x^*)|Y,x^*,X) = \mathcal{GP}(\mu,\nu)\quad$ where:


- $\mu(x^*) = k(x^*,\mathbf{X})\left(K+\sigma^2I\right)^{-1}\mathbf{f}$
- $\nu(x^*,x^*) = k(x^*,x^*)-k(x^*,\mathbf{X})\left(K+\sigma^2I\right)^{-1}k(\mathbf{X},x^*)$

```python
# mean
k_x_star_X = build_K(x_,X)
k_X_x_star = build_K(X,x_)
k_x_star_x_star = build_K(x_,x_)
Kxx = build_K(X,X)

mu = k_x_star_X @np.linalg.solve(Kxx+(noise_var)*np.eye(X.shape[0]),Y)
```

```python
# covariance matrix
cov = k_x_star_x_star - k_x_star_X @ np.linalg.solve(Kxx+(noise_var)*np.eye(X.shape[0]),k_X_x_star)
```

```python
# sample from GP posterior
num_samples = 50
samples = np.random.multivariate_normal(mean=mu.flatten(),cov=cov,size=num_samples)
```

```python
fig = plt.figure(figsize=(14,6))
_ =  plt.plot(x_,f_,'C1')
for sample in samples:
    _ = plt.plot(x_,sample,'C0',alpha=0.4)
_ = plt.plot(X,Y,'C1o')
_ =  plt.plot(x_,f_,'C1')
plt.xlabel("X")
plt.ylabel("F(X)")
_ = plt.legend(["sin(x)","samples"])
```
