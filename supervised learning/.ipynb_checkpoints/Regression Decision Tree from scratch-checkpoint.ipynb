```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# generate fake data

X = np.linspace(-1,2,200)
X = np.random.choice(X,100)
Y = np.exp(X) + np.random.rand(100)
```

```python
plt.scatter(X,Y)
plt.xlabel("$X$")
plt.ylabel("$Y$")
plt.title("$Y$ = exp($X$) + $\epsilon$")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# fact 1, dataset/ subset of size N has N-1 splits

def find_best_split(x,y):
    
    num_splits = len(x)-1
    best_split = x[1]
    best_mse = np.inf
    
    for split in range(1,num_splits):
        
        
        left_y = y[ x < x[split] ]
        
        if len(left_y) > 1:
            left_ave = np.mean(left_y)*np.ones_like(left_y)
        else:
            left_ave = left_y
        # calculate left MSE
        left_mse = np.sum((left_y-left_ave)**2)
        
        right_y = y[ x >= x[split] ]
        
        if len(right_y) > 1:
            right_ave = np.mean(right_y)*np.ones_like(right_y)
        else:
            right_ave = right_y
            
        # calculate right MSE
        right_mse = np.sum((right_y - right_ave)**2)
        
        mse = left_mse + right_mse
        #print("split ",x[split],"MSE",mse)
        if mse < best_mse:
            best_mse = mse
            best_split = x[split]
            
    #print("best_split ",x[split],"best_MSE",mse)
    return best_split, best_mse              
```

```python
# 1 find split point
def recursive_binary_splits(x,y,splits,min_leaf_samples):
    
    if len(x) <= min_leaf_samples:
        return splits.append([(np.mean(y))])
    
    split, mse = find_best_split(x,y)
    #print(splits,len(x),split)
    right_y = y[ x >= split ]
    right_x = x[ x >= split ]
    
    left_y = y[ x < split ]
    left_x = x[ x < split ]
    
    return splits.append([split,(recursive_binary_splits(left_x,left_y,splits,min_leaf_samples),recursive_binary_splits(right_x,right_y,splits,min_leaf_samples))])
        
```

```python
# crude splits
def crecursive_binary_splits(x,y,splits,min_leaf_samples):
    
    if len(x) <= min_leaf_samples:
        return np.mean(y)
    else:
    
        split, mse = find_best_split(x,y)
        #print(splits,len(x),split)
        right_y = y[ x >= split ]
        right_x = x[ x >= split ]

        left_y = y[ x < split ]
        left_x = x[ x < split ]
        splits.append([split,np.mean(left_y),np.mean(right_y)])

        crecursive_binary_splits(left_x,left_y,splits,min_leaf_samples)
        crecursive_binary_splits(right_x,right_y,splits,min_leaf_samples)
        
```

```python
# look at tree traversals, e.g. inorder binary tree traversal

tree_list = []
tree = crecursive_binary_splits(x_train,y_train,tree_list,5)
```

```python
tree_list = np.array(tree_list)

tree_list
```

```python
plt.figure(figsize=(18,18))

for l in range(1,len(tree_list)):
    p_line = tree_list[l-1]
    line = tree_list[l]
    
    plt.plot(line[0]*np.ones_like(x_train),y_train,color='r')
    #plt.plot([p_line[0],line[0]],p_line[2]*np.ones(2),color='g')
    plt.plot(x_train,line[2]*np.ones_like(x_train),color='g')
    plt.plot(x_train,line[1]*np.ones_like(x_train),color='orange')


plt.scatter(x_train,y_train,color='blue')
plt.xlabel("$X$")
plt.ylabel("$Y$")
plt.title("$Y$ = exp($X$) + $\epsilon$")
plt.show()
```
